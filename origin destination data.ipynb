```python
df=pd.read_csv('/Users/snowr/ds_python1/HOMEWORK/wk20_tableau/newYork/ny_citi_bike_ytd.csv')
```

```python
df

# for each trip: get a row for it's starting geo features and its ending geo features
# tripid, start_lat, start_long
# tripid, end_lat, end_long
# other columns can come along after the basic algorithm is done

for_dfstart = []
for_dfend = []

for index, row in df.iterrows():
    
    path_id = row['start station name'] + "|" + row['end station name']
    start_lat = row['start station latitude']
    start_long = row['start station longitude']
    start_station = row["start station name"]
    trip_duration = row['tripduration']
    for_dfstart.append({
        "lat": start_lat,
        "lng" : start_long,
        "pathid": path_id,
        "triptime": trip_duration,
        "station": start_station,
        "case": "start"
    })

start_df = pd.DataFrame.from_records(for_dfstart)
    

for index, row in df.iterrows():
    
    path_id = row['start station name'] + "|" + row['end station name']
    end_lat = row['end station latitude']
    end_long = row['end station longitude']
    end_station = row['end station name']
    trip_duration = row['tripduration']
    for_dfend.append({
        "lat": end_lat,
        "lng" : end_long,
        "pathid": path_id,
        "triptime": trip_duration,
        "station": end_station,
        "case": "end"
    })
    
end_df = pd.DataFrame.from_records(for_dfend)

#merge tables vertically
merged = pd.concat([start_df, end_df],axis=0) 
```
