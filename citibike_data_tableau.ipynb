```python
#trim the excess files NEW YORK
ny_path = '/Users/snowr/ds_python1/HOMEWORK/wk20_tableau/newYork/'
os.chdir(ny_path)
ny_files = [i for i in glob.glob('*.csv')]
ny_all_df = pd.DataFrame()
for file in ny_files:
    file = '/Users/snowr/ds_python1/HOMEWORK/wk20_tableau/newYork/' + file
    df=pd.read_csv(file) 
    ny_all_df = ny_all_df.append(df)
```

```python
columns = df.columns
```

```python
#clean up data table

#convert gender
# 0 = unknown, 1=male, 2=female
gender = {2:'female', 1:'male', 0:'unknown'}
ny_all_df['gender'] = ny_all_df['gender'].map(gender)

#generate age
def calc_age(birth_year):
    today = date.today()
    return today.year - birth_year
ny_all_df['age'] = ny_all_df['birth year'].apply(calc_age)
ny_all_df.to_csv("ny_citi_bike_ytd.csv")
# ny_all_df
```
