```python
#NEW YORK dataframe
ny_all_df = pd.DataFrame()
for file in ny_files:
    file = '/Users/snowr/ds_python1/GroupProject1/project1_data/citiBike_archive/newYork/' + file
    df=pd.read_csv(file, nrows=700)
#     print(df.columns)
    ny_all_df = ny_all_df.append(df)
```

```python
columns = df.columns
```

```python
#clean up data table

#convert gender
# 0 = unknown, 1=male, 2=female
gender = {2:'female', 1:'male', 0:'unknown'}
ny_all_df['gender'] = ny_all_df['gender'].map(gender)

#generate age
def calc_age(birth_year):
    today = date.today()
    return today.year - birth_year
ny_all_df['age'] = ny_all_df['birth year'].apply(calc_age)
ny_all_df 
```

```python
#age grouping
age_bins = [0, 20, 30, 40, 50, 60, 100]
age_groups = ['0-20', '21-30', '31-40', '41-50', '51-60', '61+']
ny_all_df['age groups'] = pd.cut(ny_all_df['age'], age_bins, labels=age_groups)
```

```python
#new limited dataframe
ny_bike_data = ny_all_df[['age', 'age groups', 'gender', 'usertype','starttime', 'start station latitude', 'start station longitude', 
                   'stoptime', 'end station latitude', 'end station longitude', 'tripduration']].copy()
ny_bike_data
```

```python
#gender count
gender = ny_bike_data['gender'].value_counts().reset_index()
gender
```
